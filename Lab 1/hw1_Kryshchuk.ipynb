```python
from IPython.display import IFrame
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('GLB.Ts_dSST.csv', header = 1)
df.drop(['J-D', 'D-N', 'DJF', 'MAM', 'JJA', 'SON'], axis = 1, inplace = True)

df.head()
```

```python
df.columns
df['sum'] = [0]*len(df)
for col in df.columns:
    if col != 'Year':
        df['sum'] += df[col]
        

df['average'] = df['sum']/12


print(df['average'])
```

```python
x = df['Year']
y = df['average']

annotations1=["   1944","   2014","   2015","   2016","   2017"]
annotations1={int(some) for some in annotations1}

for i, _ in enumerate(x):
    if x[i] in annotations1:
        plt.plot(x[i], y[i], 'go', color = '#f26c08')
        plt.annotate(x[i], (x[i], y[i]), horizontalalignment='left', size=14).set_alpha(.8)

annotations2=["  1904", "  1998"]
annotations2={int(some) for some in annotations2}

for i, _ in enumerate(x):
    if x[i] in annotations2:
        plt.plot(x[i], y[i], 'go', color = '#f26c08')
        plt.annotate(x[i], (x[i], y[i]), horizontalalignment='right', size=14).set_alpha(.8)

plt.scatter(x, y, color = '#f26c08', edgecolor = 'black', linewidth=0.7, s=80)
plt.axhline(0, color='black', linewidth=0.4)
plt.gcf().set_size_inches((17, 9))
plt.xticks(np.arange(min(x), max(x)+1, 10.0))
plt.grid(linewidth=0.3)

plt.text(2001, .260,
         "   HOTTER THAN THE\n1880-1899 AVERAGE\n\n                   COLDER", 
         va='top', fontsize=12, color='black').set_alpha(.8)

plt.text(1885, 2,
         "Annual Global Surface Temperature,\nRelative to Late 19th Century Average", 
         va='top', fontsize=15, color='black',
         fontweight=550)
```
